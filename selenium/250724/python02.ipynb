#### html 크롤링을 위한 XPATH 기본 포맷 (*형태)
- //tagname[@Attribute="value"]
- / : 절대경로를 의미 (예 : /html/body/div/a)
- // : 문서 내에서 바로 검색을 하라는 의미 (예 : //h1 -> 문서에서 바로 h1태그를 찾아와라)
- //*[@href] : 문서 내 href라는 속성을 가지고 있는 모든 태그를 다 찾아와라
- //a[@href="https://www.google.com"] : 문서 내 a 태그를 찾아오는데, href 속성값으로 구글 닷컴을 찾아와라
- (//a)[3] : 문서 내 a 태그를 찾아오는데, 세번째 a만 찾아와라 (1번부터 찾아옴)
- (//table)[last()] : 문서 내 table 태그를 찾아오는데, 마지막번째 table만 찾아와라
- (//a)[position() < 3] : 문서 내 a 태그를 찾아오는데, 앞에서부터 2개만 찾아와라
- //div[@*] : 문서 내 div 태그를 찾아오는데, 어떤 속성이던 상관없이 속성이 하나라도 있으면 찾아와라
